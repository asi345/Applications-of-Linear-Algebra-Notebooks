### a)

To find the least squares estimate, we will use the "Least Squares via QR Factorization" method, as explained in the book in Chapter 12.3. To find a least squares estimate $\tilde{x}$ for $Ax = b$, we first compute the $QR$ factorization of $A$ such that $QR = A$. Then, we multiply each side in $Ax = b$ by $Q^T$ from left, getting $Rx = Q^Tb = \tilde{b}$. Lastly, since $R$ is a upper triangular matrix, we can solve $Rx = \tilde{b}$ by Back Substitution.

Below, there are the function for Back Substitution and Least Squares via QR Factorization, respectively. They are taken from **Python Companion** of the book. They can be found in page 91.

```python
# Back Substitution
def back_sub(R, b_tilde):
    n = R.shape[0]
    x = np.zeros(n)
    for i in reversed(range(n)):
        x[i] = b_tilde[i]
        for j in range(i + 1, n):
            x[i] = x[i] - R[i, j] * x[j]
        x[i] = x[i] / R[i, i]
    return x

# Least Square via QR Factorization
def lstsqrQR(A, b):
    Q, R = np.linalg.qr(A)
    b_tilde = Q.T @ b
    x = back_sub(R, b_tilde)
    return x
```

Now, we can find the values for the vector $\hat{\theta}$. For training set, we use the first $600$ observations, so we cut the last $174$ observations from the data set. The 8 basis functions are taken from **Python Companion** exactly from page 115, and the matrix $X$ is defined accordingly. The resulting $\hat{\theta}$ is printed at the end.

```python
D = house_sales_data()
price = D['price']
area = D['area']
beds = D['beds']
condo = D['condo']
location = D['location']
N = len(price)
X = np.column_stack([np.ones(N), area, np.maximum(area-1.5,0), beds, condo, location==2, location==3, location==4])
theta_hat = lstsqrQR(X[:600], price[:600])
print("theta_hat : ", theta_hat)
```

### a)

As we remember, Tikhonov Regularized Inversion tries to minimize $||X\theta - \text{price}||^2 + \lambda ||\theta||^2$. The solution for this problem can be converted to single least square estimation by changing $X$ to $\tilde{X} = \begin{bmatrix}
X\\
\lambda \ I
\end{bmatrix}$. As explained in the book, we can find the least square estimation $\tilde{x}$ by using pseudoinverse technique since $\tilde{X}$ has linearly independent columns just as $X$. Therefore, we say that $\tilde{x} = (X^T X + \lambda I)^{-1} X^T b$. Using this derivation, we implement this technique. But first, we define a function that solves this problem using $\lambda$ as parameter. In upcoming cases, we can directly use it.

```python
def regularizedLstsqr(X, lambd, price) :
    invert_this = X.T @ X + lambd * np.identity(len(X[0]))
    theta_tilde = (np.linalg.inv(invert_this) @ X.T) @ price
    return theta_tilde
```

We now calculate $\hat{\theta}_1$ by setting $\lambda = 0.00001$.

```python
theta_hat_1 = regularizedLstsqr(X[:600], 0.00001, price[:600])
print("theta_hat_1 :", theta_hat_1)
```

Then we calculate $\hat{\theta}_2$ by setting $\lambda = 100000$.

```python
theta_hat_2 = regularizedLstsqr(X[:600], 100000, price[:600])
print("theta_hat_2 :", theta_hat_2)
```

$\hat{\theta}_1$ is clearly very close to $\hat{\theta}$. $\hat{\theta}_2$ is too far from $\hat{\theta}$. This difference is caused by the huge value of $\lambda$ for $\hat{\theta}_2$. $\lambda$ specifies how much it is important for $\hat{\theta}$ to be close $0$ vector. For $\lambda = 0.00001$, the norm of the vector is nearly ignorable. But for $\lambda = 100000$, the importance of the norm is way higher than the least square estimation for our house price regression. That is why $\hat{\theta}_2$ has really small entries compared to $\hat{\theta}$ and $\hat{\theta}_1$.

### b)

To find the best $\lambda$ for generalization, we need to calculate the RMS errors of training and test sets for each $\lambda$ value. The estimate that has the minimum test set RMS error while not much deviating from the training set RMS error gives us the best choice of $\lambda$. So we calculate these errors for $\lambda = 0, 0.001, 0.01, 0.1, 1$.

```python
for lambd in [0, 0.001, 0.01, 0.1, 1] :
    curTheta_hat = regularizedLstsqr(X[:600], lambd, price[:600])
    curTraining_rms = (sum((X[:600] @ curTheta_hat - price[:600])**2) / 600)**(0.5)
    print("RMS error on training set for lambda =", lambd, ":", curTraining_rms)
    curTest_rms = (sum((X[600:] @ curTheta_hat - price[600:])**2) / 174)**(0.5)
    print("RMS error on test set for lambda =", lambd, ":", curTest_rms, "\n")
```

The RMS errors are minimum for both training and test sets when we have $\lambda = 0$. And also, the training set and test set errors are very close to each other. Therefore, we choose $\lambda = 0$ for the best generalization.

Now, take $\lambda > 0$. We have a Tikhonov regularized inversion problem. Then, our feature vector $X_{train}$ becomes the stacked matrix $\tilde{X_{train}} = \begin{bmatrix}
X_{train}\\
\lambda \ I
\end{bmatrix}$. We may know that $X$ has linearly dependent columns, but we are sure that the stacked matrix $\tilde{X_{train}}$ has always linearly independent columns when $\lambda > 0$. This is the case since the bottom block $\lambda \ I$ has linearly independent columns. Then, this matrix is left invertible and thus we can find a solution by just 30 observations. The gram matrix of $\tilde{X_{train}}$ is $\begin{bmatrix}
X_{train} \ \lambda I
\end{bmatrix} \ \begin{bmatrix}
X_{train}\\
\lambda \ I
\end{bmatrix} = X_{train}^T X_{train} + \lambda \ I$. It is exactly the inverted part in the equation for $\hat{\theta}$. Therefore, the equation $\hat{\theta} = (X_{train}^T X_{train} + \lambda I)^{-1} X_{train}^T p$ becomes valid at these conditions and can be used directly to find $\hat{\theta}$.

For the second case, take $\lambda = 0$. The equation becomes $\hat{\theta} = (X_{train}^T X_{train})^{-1} X_{train}^T p$. We know from the above arguments that the Gram Matrix $X_{train}^T X_{train}$ is not invertible since $X$ is wide and thus has linearly dependent columns. So, the equation does not hold for $\lambda = 0$.

As an alternative, we can use the results we obtain in the previous assignment for using pseudo inverse to find least square estimation. Therefore, we can say that $\hat{\theta} = X_{train}^{\dagger} p$, with $X_{train}^{\dagger}$ being the right inverse of $X_{train}$. This calculation will give us the minimum norm solution for $\hat{\theta}$, which holds for what we want to achieve. Then, the open form of the equation will be

$\hat{\theta} = X_{train}^T (X_{train} X_{train}^T)^{-1} p$, with $p$ denoting the price vector.

First, we calculate $QR$ factorization of the stacked matrix $\begin{bmatrix}

```python
stacked = np.row_stack((A, C))
# QR factorization of the stacked matrix
Q, R = np.linalg.qr(stacked)
# Q_1 for A
Q_1 = Q[:150]
# Q_2 for C
Q_2 = Q[150:]
# QR factorization of Q_2^T
Q_tilde, R_tilde = np.linalg.qr(Q_2.T)
```

We will find the solution pair $(\tilde{c}, \tilde{z})$ using the formula $\begin{bmatrix}

\end{bmatrix} = (\begin{bmatrix}
2 A^T A & C^T\\

\end{bmatrix})^{-1} \begin{bmatrix}

\end{bmatrix}$. To apply this, we need to show that the KKT matrix is invertible. $C$ is a $6x15$ matrix, which is wide and it also has linearly independent rows. When we were applying QR factorization on stacked matrix $\begin{bmatrix}

\end{bmatrix}$, we showed that it has linearly independent columns. Therefore, KKT matrix is invertible and we can apply the given solution. We implement this idea now and obtain $[\tilde{c} \ \tilde{z}]$.

```python
KKTmatrix = np.column_stack((2 * A.T @ A, C.T))
botrow = np.column_stack((C, np.zeros((6, 6))))
KKTmatrix = np.row_stack((KKTmatrix, botrow))

rhs = np.concatenate((2 * A.T @ y, d))

cz_tilde = np.linalg.inv(KKTmatrix) @ rhs
print("[c_tilde z_tilde] :", cz_tilde)
```
